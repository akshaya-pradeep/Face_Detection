```python
# Face detection
# haar cascade model => predifined model
import cv2
```

```python
img=cv2.imread('img2_women.jpg')
img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_data=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
face_data
```

```python
faces = face_data.detectMultiScale(conv_img,minNeighbors=4) # if the face is detected min_neighbours times then only it will identify as a face
faces
#(x,y,h,w)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,255,0),4) # img name, top left point, bottom right point, color, thickness
```

```python
plt.imshow(conv_img)
```

```python
# Face detection by converting to gray scale
img1=cv2.imread('istockphoto-1368965646-612x612.jpg')
img1
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1)
```

```python
conv_img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img1)
```

```python
# convert the converted img to gray scale
img1_gray=cv2.cvtColor(conv_img1,cv2.COLOR_RGB2GRAY)
img1_gray
```

```python
plt.imshow(img1_gray)
```

```python
# read the harcascade
face_cas=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces1=face_cas.detectMultiScale(img1_gray,minNeighbors=8)
faces1
```

```python
# whether its RGB or GRAYSCALE the coordinates wont change, so we will draw rectangle in converted img
for (x,y,w,h) in faces1:
    cv2.rectangle(conv_img1,(x,y),(x+w,y+h),(0,255,0),4)
```

```python
plt.imshow(conv_img1)
```

```python
# collection of frames are called video
video1=cv2.VideoCapture(0) # to access the systems camera, to access external cam -1
while True: # starting to end of web cam
    temp,img=video1.read() # img contains each image in that reading vid, temp is always true when capturing in web cam. but when we collect data drom vid it will become false when the vid ends
    img1_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(img1_gray,minNeighbors=4)
    for (x,y,h,w) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
    cv2.imshow('video',img)
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
video1.release() # To close web cam
cv2.destroyAllWindows()
```
